```python
import pandas as pd
import numpy as np

df_sentiments = pd.read_csv('cleaned_data/chats_sentiment.csv')
```

```python
# visualize sentiment distribution (pie chart/ bar chart)
import matplotlib.pyplot as plt
import seaborn as sns

sentiment_counts = df_sentiments['sentiment'].value_counts()
sentiment_percent =  sentiment_counts / sentiment_counts.sum() * 100

# Bar chart

plt.figure(figsize=(6,4))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Label')
plt.ylabel('Number of Messages')
for i, v in enumerate(sentiment_counts.values):
    plt.text(i, v + 20, f"{v}", ha='center', fontsize=10)
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(6,6))
plt.pie(sentiment_percent, labels=sentiment_percent.index, autopct='%1.1f%%', startangle=140, colors=['#ff6b6b','#feca57','#1dd1a1'])
plt.title('Sentiment Distribution (Percentage)')
plt.show()
```

```python
# sentiment vs satisfaction analysis
df_chats = pd.read_csv('cleaned_data/chats_final.csv')

#heatmap , correlation between sentiment and satisfaction
sentiment_satisfaction = pd.crosstab(df_chats['sentiment'], df_chats['satisfaction'])
plt.figure(figsize=(8,6))
sns.heatmap(sentiment_satisfaction, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Sentiment vs Satisfaction Heatmap')
plt.xlabel('Satisfaction')
plt.ylabel('Sentiment')
plt.show()
```

```python
# correlation coefficient
sentiment_mapping = {'negative': -1, 'neutral': 0, 'positive': 1}
df_chats['sentiment_num'] = df_chats['sentiment'].map(sentiment_mapping)
correlation = df_chats['sentiment_num'].corr(df_chats['satisfaction'])
print(f'Correlation between Sentiment and Satisfaction: {correlation:.2f}')
```

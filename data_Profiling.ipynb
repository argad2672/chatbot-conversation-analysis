```python

import os

# all table names from your database
tables = [
    "action_events", "assistant_threads", "bot_user", "bots", "categories",
    "chats", "clients", "companies", "devices", "failed_jobs",
    "flows", "knowledge", "media", "menu_items", "menus",
    "migrations", "notifications", "nova_field_attachments",
    "nova_notifications", "nova_pending_field_attachments",
    "objects", "o_t_p_s", "order_items", "orders", "password_resets",
    "personal_access_tokens", "plans", "post_photos", "posts",
    "refugees", "restaurants", "states", "telescope_entries",
    "telescope_entries_tags", "telescope_monitoring", "users"
]

# make sure reports folder exists
os.makedirs("reports", exist_ok=True)

# loop over tables and generate reports
for table in tables:
    try:
        df = pd.read_sql(f"SELECT * FROM {table}", engine)
        if not df.empty:
            profile = ProfileReport(df, title=f"{table.capitalize()} Table Report")
            profile.to_file(f"reports/{table}_profile.html")
            print(f"✅ Report generated for {table}")
        else:
            print(f"⚠️ Skipped {table} (empty table)")
    except Exception as e:
        print(f"❌ Could not generate report for {table}: {e}")
```

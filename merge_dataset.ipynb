```python
import pandas as pd

# Load both datasets
df_sentiment = pd.read_csv("cleaned_data/chats_sentiment.csv")
df_satisfaction = pd.read_csv("cleaned_data/chats_with_satisfaction.csv")

# Check common column (likely 'id')
print("Common columns:", set(df_sentiment.columns).intersection(df_satisfaction.columns))

# Merge on the unique chat ID (usually 'id')
df_merged = pd.merge(df_sentiment, df_satisfaction[['id', 'satisfaction']], on='id', how='left')

# Save final unified dataset
df_merged.to_csv("cleaned_data/chats_final.csv", index=False)

print("✅ Successfully merged! File saved as cleaned_data/chats_final.csv")
print(df_merged.head())
```

```python
import pandas as pd

from sqlalchemy import create_engine

# Replace with your DB credentials
username = "root"
password = "Hawilax90"
host = "localhost"
database = "intership"

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}/{database}")

# Example: load users table
df_users = pd.read_sql("SELECT * FROM users", engine)
df_users
```

```python
# Email verification flag
df_users['is_email_verified'] = df_users['email_verified_at'].notnull().astype(int)

# Phone availability flag
df_users['has_phone'] = df_users['phone'].notnull().astype(int)

# Date of birth flag
df_users['has_dob'] = df_users['dob'].notnull().astype(int)

# Profile photo flag
df_users['has_photo'] = df_users['photo'].notnull().astype(int)
```

```python
df_users.to_csv('cleaned_data/users_clean.csv', index=False)
```

```python
df_chats = pd.read_sql("SELECT * FROM chats", engine)
df_chats
```

```python
import numpy as np

df_chats['type'] = np.where(df_chats['type'].isnull() & df_chats['image_url'].notnull(), 'image', df_chats['type'])
df_chats['type'] = np.where(df_chats['type'].isnull() & df_chats['audio_url'].notnull(), 'audio', df_chats['type'])

# If still missing after checks, assume 'text'
df_chats['type'] = df_chats['type'].fillna('text')

# Step 2: Create helper flags
df_chats['has_image'] = df_chats['image_url'].notnull().astype(int)
df_chats['has_audio'] = df_chats['audio_url'].notnull().astype(int)

```

```python
df_chats.to_csv('cleaned_data/chats_clean.csv', index=False)
```
